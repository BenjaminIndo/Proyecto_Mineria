# Exploración del Dataset

Cargar el csv
=============
Carguemos el csv y exploremos un poco de info sobre este

```python
import pandas as pd
df = pd.read_csv('./data/Students_Grading_Dataset.csv')
df.info()
```

```python
shape = df.shape
print(f"El dataset tiene {shape[0]} filas y {shape[1]} columnas")
```

```python
df.describe(include='all')
```

```python
import matplotlib.pyplot as plt


asistencia=df['Attendance (%)']
plt.hist(asistencia, bins=range(0, 100), edgecolor='black')
plt.xlabel('Asistencia')
plt.ylabel('Frecuencia')
plt.title('Distribución de Asistencia')
plt.show()
```
